```python
# wodpyがインストールされている前提
from wodpy import wod
# ファイルを開く
fid = open("CTDO3110")
# ファイルを開いて一つのプロファイルをprofilesという配列に追加していく
profiles=[]
while True:
    profiles.append(wod.WodProfile(fid))
    if profiles[-1].is_last_profile_in_file(fid):
    # profilesがファイルで最後のプロファイルだった場合ループから抜け出す。
        break
```

```python
# 最新10件分のプロファイルの年と緯度経度をprintする。

for profile in profiles[-10:]:
    print([profile.year(), profile.latitude(),profile.longitude()])
```

```python
# 最新のもののdataframe(df)を表示
profile = profiles[-1]
profile.df()
```
